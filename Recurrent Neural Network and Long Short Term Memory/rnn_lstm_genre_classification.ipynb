```python
import sys
import os
sys.path.append(os.path.abspath('..'))

from data_handler import DataUtils
from model_builder import ModelBuilder
from sklearn.model_selection import train_test_split
import keras
from keras import layers
from keras import ops
import numpy as np
import matplotlib.pyplot as plt

DATASET_PATH = "../data.json"
data_util = DataUtils(DATASET_PATH)
inputs, targets = data_util._get_inputs_and_targets()
print(inputs.shape)
x_train, x_validation, x_test, y_train, y_validation, y_test = data_util.prepare_data(test_size=.2,validation_size=.1, model_type="LSTM")
```

```python
# Drop Third Dimension In "prepare_data", pass parameter
# Change Input Shape 2 Dimension, Drop the "depth" of CNN
# drop all convolution layer
# same output layer
# 1st LSTM layer in builder model 64 unit / neuron, inputshape, return_sequences=true
# return_sequences=true => sequences to sequences layer
# 2nd LSTM layer units = 64 // only units since return_sequences=false => is default which is a sequences to vector layer
# 3rd Layer dense, units 64, activation = relu
# 4th Layer dropout, 0.3
# last is output same soft max dense etc....
layer_config = [
    {'type': 'lstm', 'units': 64, 'return_sequences': True}, # Sequence to Sequence approach
    {'type': 'lstm', 'units': 64, 'return_sequences': False}, # Sequence to Vector approach
    
    {'type': 'dense', 'units': 64, 'activation': 'relu'},
    {'type': 'dropout', 'rate': 0.3}
]

input_shape = x_train.shape[1], x_train.shape[2]
print(input_shape)
num_classification = 10
model_builder = ModelBuilder(input_shape, num_classification, layers=layer_config)
model = model_builder._build_model(learning_rate=0.0001)
# Train
history = model.fit(x_train, y_train, validation_data=(x_validation, y_validation), batch_size=32, epochs=50)
# Evaluate
test_error, test_accuracy = model.evaluate(x_test, y_test, verbose=1)
print("Accuracy on Test Set: ", test_accuracy)
print("Error on Test Set: ", test_error)
```
